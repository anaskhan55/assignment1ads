```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('E:/ANAS/data.csv', parse_dates=['Date'], index_col=['Date'])
```

```python
print(data.columns)
```

```python
def plot_monthly_high_temps(data):
    # group data by month and calculate the average daily high temperature
    monthly_highs = data.groupby(pd.Grouper(freq='M')).mean()['High Temp (°F)']
    
    # set the colors for the bars
    colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan', 'm', 'y']
    
    # plot the data as a bar graph
    plt.figure(figsize=(12,6))
    plt.bar(monthly_highs.index.month_name(), monthly_highs.values, color=colors)
    
    # set the axis labels and title
    plt.xlabel('Month')
    plt.ylabel('Average daily high temperature (°F)')
    plt.title('Average daily high temperature by month')
    
    # show the plot
    plt.show()

# call the function with the data as input
plot_monthly_high_temps(data)
```

```python
def scatter_plot(df, x_col, y_col, x_label='', y_label='', title=''):
    plt.figure(figsize=(12, 6))
    plt.scatter(df[x_col], df[y_col])
    plt.xlabel(x_label if x_label else x_col)
    plt.ylabel(y_label if y_label else y_col)
    plt.title(title if title else f'Scatter plot of {x_col} vs {y_col}')
    plt.show()
scatter_plot(data, 'Precipitation', 'Total', 'Precipitation (inches)', 'Total traffic', 'Daily total traffic vs precipitation')
```

```python
def line_graph(data, columns):
    # create a new DataFrame with only the selected columns
    bridge_data = data[columns]

    # create x-axis data
    x = np.linspace(0, 10, len(bridge_data))

    # create y-axis data for each column
    y1 = bridge_data[columns[0]]
    y2 = bridge_data[columns[1]]

    # plot the data as a line plot
    plt.plot(x, y1, label=columns[0])
    plt.plot(x, y2, label=columns[1])

    # set the axis labels and title
    plt.xlabel('Date')
    plt.ylabel('Number of Vehicles')
    plt.title('Manhattan Bridge and Williamsburg Bridge Traffic')

    # add legend
    plt.legend()

    # show the plot
    plt.show()
columns = ['Manhattan Bridge', 'Williamsburg Bridge']
line_graph(data, columns)
```
